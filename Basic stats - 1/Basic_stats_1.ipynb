```python
# Step 1: Import required libraries
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
# Step 2: Create a DataFrame with the durability data (in millions of characters)
durability_values = {
    'Durability': [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                   1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
}
df_durability = pd.DataFrame(durability_values)
sample_size = len(df_durability['Durability'])
sample_mean = df_durability['Durability'].mean()
sample_std_dev = df_durability['Durability'].std(ddof=1)  # sample std dev

sample_mean, sample_std_dev, sample_size
```

## Confidence Interval using Sample Std Dev (t-distribution)

When the population standard deviation is unknown and sample size is small (n < 30), we use the **t-distribution** for constructing the confidence interval.

```python
confidence_level = 0.99
t_confidence_interval = stats.t.interval(
    confidence_level,
    df=sample_size - 1,
    loc=sample_mean,
    scale=sample_std_dev / np.sqrt(sample_size)
)
t_confidence_interval
```

## Confidence Interval using Known Population Std Dev (z-distribution)

If the population standard deviation is known (assumed to be 0.2 million characters), we can use the **z-distribution** for a more precise confidence interval.

```python
population_std_dev = 0.2
z_confidence_interval = stats.norm.interval(
    confidence_level,
    loc=sample_mean,
    scale=population_std_dev / np.sqrt(sample_size)
)
z_confidence_interval
```

## Summary
- **t-distribution CI (sample std dev):** `t_confidence_interval`
- **z-distribution CI (known std dev = 0.2):** `z_confidence_interval`

These intervals allow us to estimate the true average durability of print-heads with 99% confidence.
